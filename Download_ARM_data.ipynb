Download ARM data using ftp

```python
import xarray as xr
import matplotlib.pyplot as plt
from glob import glob
import os
from ftplib import FTP
```

Open files using ftp

```python
ftp_address='ahmedf2/228707' #arm-CACTI sounding
ftp_address='ahmedf2/228875' # arm-TWPC1C3
ftp_address='ahmedf2/228873' # arm-TWPC2
ftp_address='ahmedf2/228910' # ARM-GoAmazon
ftp_address='ahmedf2/228908' # ARM-CACTI
ftp_address='ahmedf2/228909' # ARM-SGP

ftp = FTP('ftp.archive.arm.gov')   # connect to host, default port
ftp.login()
ftp.cwd(ftp_address) # sounding
listing = []
ftp.retrlines("LIST", listing.append)
```

```python
for name in listing[2:]:
    words = name.split(None, 8)
    filename = words[-1].lstrip()
    print(filename)
    local_filename = os.path.join(local_dir, filename)
    lf = open(local_filename, "wb")
    ftp.retrbinary("RETR " + filename, lf.write, 8*1024)
    lf.close()
ftp.quit() 
```
